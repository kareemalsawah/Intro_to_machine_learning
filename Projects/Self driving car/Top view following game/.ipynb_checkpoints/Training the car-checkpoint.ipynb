```python
import numpy as np
```

```python
data_collected = np.load("data_saved.npy",allow_pickle=True)
```

```python
data_collected.shape
```

```python
data_X = []
data_y = []
for arr in data_collected:
    data_X_arr = []
    data_X_arr.append(arr[0])
    data_X_arr.append(arr[1])
    data_X_arr.append(arr[2])
    data_y_arr = []
    for i in range(3,6):
        if arr[i]:
            data_y_arr.append(1)
        else:
            data_y_arr.append(0)
    data_X.append(data_X_arr)
    data_y.append(data_y_arr)
data_y
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(data_X, data_y)
```

```python
np.save("coefs",regr.coef_)
np.save("intercepts",regr.intercept_)
```

```python
regr.predict([[1,1,1]])
```

```python
y_index = 2
sum(regr.coef_[y_index])+regr.intercept_[y_index]
```

```python
regr.score(data_X,data_y)
```
